# Prelim Metric / Rank



```python
bowling_records['prelim_metric'] = np.round((bowling_records['average'] * bowling_records['economy'])/6 , 3)
```

```python
bowling_records.loc[((bowling_records['prelim_metric'] < 30) & (bowling_records.wickets >= 25)), 'pre-rank'] = 'Best'
bowling_records.loc[((bowling_records['prelim_metric'] < 30) & (bowling_records.wickets < 25)), 'pre-rank'] = 'Good'
bowling_records.loc[((bowling_records['prelim_metric'] < 40) & (bowling_records['prelim_metric'] >30)), 'pre-rank'] = 'Good'
bowling_records.loc[((bowling_records['prelim_metric'] <65 ) & (bowling_records['prelim_metric'] >40)), 'pre-rank'] = 'Average'
bowling_records.loc[((bowling_records['prelim_metric'] > 65)), 'pre-rank'] = 'Poor'
bowling_records.loc[(bowling_records.wickets < 5), 'pre-rank'] = 'Poor'

```

# Feature Scaling

#### Normal Data

```python
wcss = []
for i in range(1,11,1):
  model = KMeans(i).fit(bowler_for_clustering)
  wcss.append(model.inertia_)
plt.plot(np.arange(1,11,1),wcss)
plt.show()
```

```python
plt.figure(figsize=(17,8)) 
sns.boxplot(data=bowler_for_clustering)
plt.show()
```

```python
fig = plt.figure(figsize = (15,10))
ax = fig.gca()
bowler_for_clustering.hist(ax=ax)
plt.show()
```

```python
range_n_clusters = [2, 3, 4, 5, 6,7,8,9,10]

for n_clusters in range_n_clusters:

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(bowler_for_clustering)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(bowler_for_clustering, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )
```

```python
corr_normal = bowler_for_clustering.corr()
plt.figure(figsize = (10,6))
sns.heatmap(corr_normal,annot=True)
```

#### Formula based *Normalisation*

```python
bowler_nn = bowler_for_clustering.copy()

bowler_nn['average'] = ((np.max(bowler_nn['average'])) - bowler_nn['average']) / (np.max(bowler_nn['average']) - np.min(bowler_nn['average']))
bowler_nn['strike_rate'] = ((np.max(bowler_nn['strike_rate'])) - bowler_nn['strike_rate']) / (np.max(bowler_nn['strike_rate']) - np.min(bowler_nn['strike_rate']))
bowler_nn['economy'] = ((np.max(bowler_nn['economy'])) - bowler_nn['economy']) / (np.max(bowler_nn['economy']) - np.min(bowler_nn['economy']))
bowler_nn['runs_index'] = ((np.max(bowler_nn['runs_index'])) - bowler_nn['runs_index']) / (np.max(bowler_nn['runs_index']) - np.min(bowler_nn['runs_index']))


bowler_nn['balls_bolwed_per_innings'] = (bowler_nn['balls_bolwed_per_innings'] - np.min(bowler_nn['balls_bolwed_per_innings'])) / (np.max(bowler_nn['balls_bolwed_per_innings']) - np.min(bowler_nn['balls_bolwed_per_innings']))
bowler_nn['wicket_index'] = (bowler_nn['wicket_index'] - np.min(bowler_nn['wicket_index'])) / (np.max(bowler_nn['wicket_index']) - np.min(bowler_nn['wicket_index']))
bowler_nn['big_impact_index'] = (bowler_nn['big_impact_index'] - np.min(bowler_nn['big_impact_index'])) / (np.max(bowler_nn['big_impact_index']) - np.min(bowler_nn['big_impact_index']))
bowler_nn['short_impact_index'] = (bowler_nn['short_impact_index'] - np.min(bowler_nn['short_impact_index'])) / (np.max(bowler_nn['short_impact_index']) - np.min(bowler_nn['short_impact_index']))


```
