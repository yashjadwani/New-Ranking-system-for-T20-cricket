##### Importing Top 10 ICC All time All rounder after pre processing in Excel.

```python
top10_ICC_AR = pd.read_csv('/content/drive/MyDrive/Diss/Code/CSV/preProcesing_allRounder/ICC_top10_allrounders.csv')
top10_ICC_AR['ICC_rank'] = top10_ICC_AR.name.apply(lambda x: top10AllTime_allRounders.get(x))
```

```python
top10_ICC_AR
```

```python
top10_ICC_AR['RBML_rank'] = top10_ICC_AR['combined_rbml_Score'].rank(method='max',numeric_only=True,ascending=False)
top10_ICC_AR['PCA_rank'] = top10_ICC_AR['combined_pca_score'].rank(method='max',numeric_only=True,ascending=False)
top10_ICC_AR['hussey_rank'] = top10_ICC_AR['hussey_index(bat)'].rank(method='max',numeric_only=True,ascending=False)
top10_ICC_AR
```

```python
convert_dtypes = {'RBML_rank':int,'PCA_rank':int,'hussey_rank':int}
top10_ICC_AR=top10_ICC_AR.astype(convert_dtypes)
```

```python
cols = top10_ICC_AR.columns.tolist()
cols = ['ICC_rank','name','country','batting_pre_rank','bowling_pre_rank','batting_role','bowling_role','combined_rbml_Score','RBML_rank','combined_pca_score','PCA_rank','hussey_index(bat)','hussey_rank']
top10_ICC_AR = top10_ICC_AR[cols]
```

```python
top10_ICC_AR.sort_values('ICC_rank')
```

```python
filepath_allrounder_icc = Path('ICC_allTime_top_10_allRounder.csv') 
top10_ICC_AR.sort_values('ICC_rank').to_csv(filepath_allrounder_icc,index=False)
```
